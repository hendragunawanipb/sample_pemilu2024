# Tambahkan beberapa kolom
* `No`
* `suara01` dari kolom `chart.100025`
* `suara02` dari kolom `chart.100026`
* `suara03` dari kolom `chart.100027`
* `totalSah` dari kolom `administrasi.suara_sah`
* `totalSuara` dari kolom `'chart.100025' + 'chart.100026' + 'chart.100027'`
* `perbedaanSuara` dari kolom `administrasi.suara_sah - totalSuara`
* `perbedaan` dari `perbedaanSuara`
  * jika nilainya 0  `perbedaan` = `False`,
  * jika nilainya tidak 0 `perbedaan` = `True` (bisa jadi nilainya lebih besar/positif atau lebih kecil/negatif)

````python
# prompt: add column "totalChart" into combined_df_filtered based on sum of child chart
combined_df_filtered['No'] = np.arange(len(combined_df_filtered))
combined_df_filtered['suara01'] = combined_df_filtered['chart.100025']
combined_df_filtered['suara02'] = combined_df_filtered['chart.100026']
combined_df_filtered['suara03'] = combined_df_filtered['chart.100027']
combined_df_filtered['totalSah'] = combined_df_filtered['administrasi.suara_sah']
combined_df_filtered['totalSuara'] = combined_df_filtered[['chart.100025', 'chart.100026', 'chart.100027']].sum(axis=1)
combined_df_filtered['perbedaanSuara'] = combined_df_filtered['administrasi.suara_sah'] - combined_df_filtered['totalSuara']
combined_df_filtered['url'] =  'https://pemilu2024.kpu.go.id/pilpres/hitung-suara/' + combined_df_filtered['KodeTPS'].str[:2] +'/'+ combined_df_filtered['KodeTPS'].str[:4] +'/'+ combined_df_filtered['KodeTPS'].str[:6] +'/'+ combined_df_filtered['KodeTPS'].str[:10] + '/'+ combined_df_filtered['KodeTPS']



def make_clickable(val):
    return '<a href="{}">{}</a>'.format(val,val)

df.style.format(make_clickable)

def cekPerbedaan(row):
    if row['perbedaanSuara'] == 0:
        val = False
    else:
        val = True
    return val

combined_df_filtered['perbedaan'] = combined_df_filtered.apply(cekPerbedaan,axis=1)
#combined_df_filtered
````
